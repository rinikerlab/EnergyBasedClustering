```python
from sklearn import datasets
from sklearn.neighbors import KernelDensity

import numpy as np
```

```python
n_samples = 1000

samples, labels = datasets.make_moons(n_samples=n_samples, noise=0.05)
samples, labels = datasets.make_blobs(n_samples=n_samples, random_state=8)
kde = KernelDensity(bandwidth=0.1).fit(samples)
```

```python
scores = kde.score_samples(samples) 
energies = np.exp(-scores)
```

```python
ebc = EBC(n_clusters=3, use_sparse=False) 
ebc.fit(samples, energies)
ebc.show()
```
