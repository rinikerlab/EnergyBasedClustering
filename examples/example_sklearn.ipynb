```python
from sklearn import datasets
from sklearn.neighbors import KernelDensity

import numpy as np
```

```python
n_samples = 1000

samples, labels = datasets.make_moons(n_samples=n_samples, noise=0.05)
samples, labels = datasets.make_blobs(n_samples=n_samples, random_state=8)
samples, labels = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)
#samples, _ = datasets.make_blobs(n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=170) # 1e-1
#X, labels = datasets.make_blobs(n_samples=n_samples, random_state=170)
#transformation = [[0.6, -0.6], [-0.4, 0.8]]
#samples = np.dot(X, transformation)

kde = KernelDensity(bandwidth=0.1).fit(samples)
```

```python
scores = kde.score_samples(samples) 
energies = np.exp(-scores)
```

```python
ebc = EBC(n_clusters=2, use_sparse=True) 
ebc.fit(samples, energies)
ebc.show();
```
